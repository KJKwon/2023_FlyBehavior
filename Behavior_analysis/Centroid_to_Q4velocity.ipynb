# Calculate centroid velocity

```python
##DGRPXXX_[#]_centroid.txt
#Centroid file
#time	x	y
#60.001515	583.41	335.90
#60.268336	582.99	337.26

#Using centroid position profile produced with position_to_centroid.py
os.chdir('/home/kujin/Documents/Project/Social_Drosophila.2022/Analysis_centroid/')
dir_list = sorted(os.listdir('.'))
DGRP_Arena2velo = dict()
#Centroid velocities of DGRP arena are saved in DGRP_Arena2velo
for f_cent_nm in dir_list:
    if f_cent_nm.endswith('centroid.txt'):
        DGRP_num = f_cent_nm.rstrip('_centroid.txt')
        f_cent = open(f_cent_nm,'r')
        f_cent.readline()
        line_no = 0
        for line in f_cent:
            tokens = line.strip().split('\t')
            cent_time = float(tokens[0])
            cent_x = float(tokens[1])
            cent_y = float(tokens[2])
            #Calculate centorid velocity by each time of frame interval
            if line_no != 0:
                cent_dist = ((cent_x-bf_cent_x)**2 + (cent_y-bf_cent_y)**2)**(1/2)
                time_lapse = cent_time - bf_cent_time
                cent_velo = cent_dist/time_lapse
                if DGRP_num not in DGRP_Arena2velo:
                    DGRP_Arena2velo[DGRP_num] = []
                DGRP_Arena2velo[DGRP_num].append(cent_velo)
            line_no += 1
            bf_cent_time = float(cent_time)
            bf_cent_x = float(tokens[1])
            bf_cent_y = float(tokens[2])
```

# Calculate centroid velocity IQR in 4th qunatile of recording time

```python
import numpy as np
import math

DGRP_Arena_list = sorted(DGRP_Arena2velo.keys())
DGRP_Arena2IQR = dict()

for DGRP_Arena in DGRP_Arena_list:
    cent_velo = DGRP_Arena2velo[DGRP_Arena]
    ##Only centroid velocity about 7.5min to 10min considered (end 25% of total recording time)
    velo_Q4 = np.array(cent_velo[round(len(cent_velo)*0.75):])
    velo_Q4_1Q = np.quantile(velo_Q4, 0.25)
    velo_Q4_3Q = np.quantile(velo_Q4, 0.75)
    velo_IQR = velo_Q4_3Q - velo_Q4_1Q
    print(DGRP_Arena, velo_IQR)
```
